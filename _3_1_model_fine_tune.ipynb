```python
# Determine the number of existing and new data points
num_existing_data = len(data['train']['sentence'])
num_new_data = len(new_train_data['sentence'])

# Insert the new data randomly into the existing dataset
combined_data = {
    "sentence": [],
    "sentiment": [],
    "topic": [],
}

for i in range(num_existing_data + num_new_data):
    if i < num_existing_data:
        combined_data['sentence'].append(data['train']['sentence'][i])
        combined_data['sentiment'].append(data['train']['sentiment'][i])
        combined_data['topic'].append(data['train']['topic'][i])
    else:
        random_index = random.randint(0, i)  # Insert randomly within the existing and new data
        combined_data['sentence'].insert(random_index, new_train_data['sentence'][i - num_existing_data])
        combined_data['sentiment'].insert(random_index, new_train_data['sentiment'][i - num_existing_data])
        combined_data['topic'].insert(random_index, new_train_data['topic'][i - num_existing_data])

data['train'] = Dataset.from_dict(combined_data)
```

```python
from transformers import RobertaForSequenceClassification, AutoTokenizer

model = RobertaForSequenceClassification.from_pretrained("wonrax/phobert-base-vietnamese-sentiment")
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    precision = precision_score(labels, predictions, average='weighted', zero_division=1)
    recall = recall_score(labels, predictions, average='weighted')
    acc = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='weighted')

    return {
        "accuracy": acc,
        "f1": f1,
        "precision": precision,
        "recall": recall,
    }
```

```python
tokenized_datasets.shape
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
predictions = trainer.predict(tokenized_datasets["test"])

y_pred = np.argmax(predictions.predictions, axis=1)
y_actual = predictions.label_ids
# Print all predicted and actual labels
for i in range(len(y_pred)):
    print(f"Example {i + 1} - Predicted: {y_pred[i]}, Actual: {y_actual[i]}")
predictions
```
